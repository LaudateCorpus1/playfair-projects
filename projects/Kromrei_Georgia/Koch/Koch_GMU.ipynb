```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import dateutil.parser
```

```python
df = pd.read_csv("Koch_GMU_Yearly.csv")
```

```python
df.columns
df.dtypes
df.index=df['Year']
```

```python
#What was the percent increase from year to year
def percent_increase(index):
    return (df.ix[index+1].sum()-df.ix[index].sum())/df.ix[index].sum()*100

df_before_2014 = df[df['Year'] < 2014]
df_before_2014['Year'].apply(percent_increase)

(df.ix[2014].sum()-df.ix[2012].sum())/df.ix[2012].sum()*100
```

```python
fig, ax = plt.subplots()

ax = df.plot.bar(stacked='True', alpha = .5, lw=0, figsize = (10,8))
ax.set_clip_on(False)
ax.axis([2005,2014,0,15000000])
ax.set_xlabel("Year")
ax.set_ylabel("Donation")

ax.tick_params(which= 'both', top= 'off', bottom = 'off', left = 'off', right = 'off')
ax.set_axisbelow(True)
ax.grid(linestyle=':', linewidth='0.4', color='grey')
ax.set_title("Koch Donations to George Mason University, by Institute, 2005-2014")
ax.set_ylim(0,18000000)
df.plot.bar(stacked="True", lw=0, ax=ax)
plt.savefig("Koch_Donations_by_Institute_Stacked.pdf",transparent=True)
```
