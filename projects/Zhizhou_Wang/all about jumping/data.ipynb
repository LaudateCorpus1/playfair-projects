```python
import pandas as pd
from datetime import datetime
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('default')
import dateutil.parser
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
pvm = pd.read_excel("pvmen.xlsx")
```

```python
def get_year(d):
    d = str(d)
    return datetime.strptime(d,'%Y')
```

```python
datetime.strptime('2016', '%Y')
```

```python
pvm['Year'] = pvm['Year'].apply(get_year)
```

```python
pvm
```

```python
pvm1 = pvm
```

```python
pvm1.index=pvm1['Year']
```

```python
pvm1.plot()
```

```python
pvm1.sort_values(by='Score',ascending=False)
```

```python
ax = pvm1.plot(kind='bar',figsize=[20,10])
ax

# plt.savefig('pvm.pdf',transparent=True)
```
