## Chart 1: New York City Rape

Data from: 
- http://www.criminaljustice.ny.gov/crimnet/ojsa/crimereporting/2015-Preliminary-Crime-Release.pdf
- https://ucr.fbi.gov/crime-in-the-u.s/2013/crime-in-the-u.s.-2013/violent-crime/rape
- http://www.criminaljustice.ny.gov/crimnet/ojsa/indexcrimes/county_totals.htm#t

Six questions for the data: 

1. Are there differences between boroughs?
2. What's the difference between years?
3. What was the greatest number reported? 
4. The lowest?
5. What definitions do they use? 
6. Who/how is the data collected and reported?

Design Decisions: 
- Tried to make vertical charts instead (with year on the x-axis), but went with horizontal charts for aesthetic reasons. 
- All text, gridlines and most bars in shades of grey. The bar I'd like to call attention to is in maroon. 

## Chart 2: Jail Population by County 

Data from: 
- http://www.census.gov/en.html 
- https://data.ny.gov/Public-Safety/Jail-Population-By-County-Beginning-1997/nymx-kgkn

Six questions for the data: 

1. How has the jail population changed generally over time?
2. Are there any outlier counties with abnormally high or low populations?
3. When was jail population the highest? 
4. The lowest? 
5. What definitions do they use? 
6. Who/how is the data collected and reported?

Design Decisions: 
- Again, all text, points, grid lines in shades of grey or black. 
- Data points I'd like to call attention to in maroon. 
- Wanted to connect the lines (line chart) to make the trend for every jail more clear, but thought it would look too cluttered. Might be better using D3 so you could rollover each line for more info. 

## Problems Encountered

## Work: Chart 1

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
NY_df = pd.read_excel('EditedNY.xlsx')
NY_df.head()
```

```python
NY_df.sort_values('Total', ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(10,5))

NY_df.groupby('Year')['Rape'].median().plot(kind='barh', ax=ax, linewidth=0)
ax.set_ylabel("")
ax.grid(color='darkgrey', linestyle=':', linewidth=1)
ax.set_axisbelow(True)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.tick_params(
    which='major',
    top='off',
    left='off',
    right='off',
    bottom='off',
    labeltop='off',
    labelbottom='on')

ax.set_xlim((0,600))
plt.savefig("nyc.pdf", transparent=True) 
```

## Work: Chart 2

```python
df = pd.read_csv("Jail_Population_By_County__Beginning__1997.csv")
df.head(2)
```

```python
census_df = df.groupby('Census')
census_df.head()
```

```python
df.sort_values('Census')
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.set_prop_cycle('color', ['grey'])

for category, group in df.groupby('Facility Name'):
    ax.plot(group['Year'], group['Census'], marker='o', linestyle='', label=category, markeredgewidth=0)
ax.set_xlabel("")

ax.grid()
ax.set_axisbelow(True)
ax.grid(linestyle=':', linewidth='0.5', color='darkgrey')
ax.minorticks_on()
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='darkgrey')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.tick_params(which='both', 
                top='off',
                left='off', 
                right='off',  
                bottom='off')

ax.set_xlim((1996,2017))
ax.set_ylim(-100,2000)


plt.savefig("jails.pdf", transparent=True) 
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.set_prop_cycle('color', ['grey'])

for category, group in df.groupby('Facility Name'):
    ax.plot(group['Year'], group['Census'], marker='', linestyle='-', label=category, markeredgewidth=0.2)

ax.set_xlabel("")
ax.grid()
ax.set_axisbelow(True)
ax.grid(linestyle=':', linewidth='0.5', color='darkgrey')
ax.minorticks_on()
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='darkgrey')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.tick_params(which='both', 
                top='off',
                left='off', 
                right='off',  
                bottom='off')

ax.set_xlim((1996,2017))
ax.set_ylim(-100,2000)


plt.savefig("jails2.pdf", transparent=True) 
```
