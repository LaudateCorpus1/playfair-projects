```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
import matplotlib
matplotlib.rc('pdf', fonttype=42)
```

```python
df = pd.read_csv('Inmates_Under_Custody__Beginning_2008.csv')
df.head(1)
```

```python
df['County of Indictment'].value_counts().sort_values(ascending=True).plot(kind='barh', width=.9, figsize=(7,18))
plt.savefig("counties.pdf", transparent=True) 
```

```python
fourteen_df = df[df['Snapshot Year'] == 2014]
```

```python
fig, ax = plt.subplots(figsize=(15,2))
fourteen_df['Gender'].value_counts().sort_values(ascending=False).plot(kind='barh', figsize=(15,2))
plt.savefig("2014gender.pdf", transparent=True) 
```

```python
male2 = fourteen_df[fourteen_df['Gender'] == 'MALE']
female2 = fourteen_df[fourteen_df['Gender'] == 'FEMALE']
```

```python
male2_df = male2['Current Age'].sort_values(ascending=False)
male2_df.tail(5)
```

```python
male2_df.head(1)
```

```python
female2_df = female2['Current Age'].sort_values(ascending=False)
female2_df.tail(5)
```

```python
male2['Most Serious Crime'].value_counts().head(10).sort_values(ascending=True).plot(kind='barh', color='blue', figsize=(10,5))
female2['Most Serious Crime'].value_counts().head(10).sort_values(ascending=True).plot(kind='barh', color='red', figsize=(10,5))
plt.savefig("2014mostseriouscrimes.pdf", transparent=True) 
```

```python
fig, ax = plt.subplots(figsize=(30,30))
male2['Current Age'].value_counts().sort_index().plot(kind='bar', figsize=(30,20), width=1, color='blue')
plt.savefig("2014male.pdf", transparent=True) 
```

```python
fig, ax = plt.subplots(figsize=(20,10))
female2['Current Age'].value_counts().sort_index().plot(kind='bar', figsize=(20,10), width=1, color='red')
plt.savefig("2014female.pdf", transparent=True) 
```

```python
fig, ax = plt.subplots(figsize=(20,10))
male2['Current Age'].value_counts().sort_index().plot(kind='bar', width=1, color='blue')
female2['Current Age'].value_counts().sort_index().plot(kind='bar', width=1, color='red')
plt.savefig("2014femaleandmale.pdf", transparent=True) 
```
