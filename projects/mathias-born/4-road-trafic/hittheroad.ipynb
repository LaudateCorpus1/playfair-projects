```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
plt.style.use('fivethirtyeight')
```

```python
# Importing the data.

df_1 = pd.read_csv('data/1.csv', delimiter=';', encoding = 'ISO-8859-1')
df_2 = pd.read_csv('data/2.csv', delimiter=';', encoding = 'ISO-8859-1')
df_3 = pd.read_csv('data/3.csv', delimiter=';', encoding = 'ISO-8859-1')
df_4 = pd.read_csv('data/4.csv', delimiter=';', encoding = 'ISO-8859-1')
df_5 = pd.read_csv('data/5.csv', delimiter=';', encoding = 'ISO-8859-1')
df_6 = pd.read_csv('data/6.csv', delimiter=';', encoding = 'ISO-8859-1')
df_7 = pd.read_csv('data/7.csv', delimiter=';', encoding = 'ISO-8859-1')
df_8 = pd.read_csv('data/8.csv', delimiter=';', encoding = 'ISO-8859-1')
df_9 = pd.read_csv('data/9.csv', delimiter=';', encoding = 'ISO-8859-1')
df_10 = pd.read_csv('data/10.csv', delimiter=';', encoding = 'ISO-8859-1')
df_11 = pd.read_csv('data/11.csv', delimiter=';', encoding = 'ISO-8859-1')
df_12 = pd.read_csv('data/12.csv', delimiter=';', encoding = 'ISO-8859-1')
df_13 = pd.read_csv('data/13.csv', delimiter=';', encoding = 'ISO-8859-1')
```

```python
# Merge the data. 
df = pd.merge(df_1, df_2, how='outer')
df = pd.merge(df, df_3, how='outer')
df = pd.merge(df, df_4, how='outer')
df = pd.merge(df, df_5, how='outer')
df = pd.merge(df, df_6, how='outer')
df = pd.merge(df, df_7, how='outer')
df = pd.merge(df, df_8, how='outer')
df = pd.merge(df, df_9, how='outer')
df = pd.merge(df, df_10, how='outer')
df = pd.merge(df, df_11, how='outer')
df = pd.merge(df, df_12, how='outer')
df = pd.merge(df, df_13, how='outer')

# Reorder the columns.
df = df[['Kanton', 'Karosserie', 'Treibstoff', 'Anzahl Sitzplätze', 'Antrieb',
       'Getriebe', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997',
       '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006',
       '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
```

## Import Census data

```python
# Let's import census data. We have to do it in two steps and we have to do some data wrangling since the data is not available for the full period. 
df_census1 = pd.read_csv('data/wohnbevölkerung.csv', delimiter=';', encoding = 'ISO-8859-1')
del df_census1['Demographische Komponente']
df_census1.index = df_census1['Wohnkanton']
del df_census1['Wohnkanton']
df_census1.index.name = 'Kanton'
df_census1 = df_census1.drop(df_census1.index[[0]])
```

```python
# I am getting rid of the data for the year 2010 now, since it is included in the second dataset. 
del df_census1['2010']
df_census1.head()
```

```python
df_census2 = pd.read_csv('data/wohnbevölkerung2.csv', delimiter=',', encoding = 'ISO-8859-1')
df_census2 = df_census2.pivot(index='Kanton', columns='Jahr', values='Ständige und nichtständige Wohnbevölkerung')
df_census2.head()
```

```python
# We'll have to add some description because the column names should be different for a merge. 
df_census = pd.concat([df_census1, df_census2], axis=1)
df_census = df_census.add_suffix('_pop')
df_census.index.name = 'Kanton'
df_census.head()
```

```python
# In which canton are the most 4-wheelers per capita?
df_4wheel = df[df['Antrieb'] == 'Allrad'].groupby(by='Kanton').sum()
df_4wheel = df_4wheel.join(df_census)
df_4wheel
```

## Comparison between Cantons

```python
# Drop the row 'Bund'. It's causing troubles because the 'Bund' doesn't have inhabitants...
df_4wheel = df_4wheel.drop(['Bund'])
```

```python
# Let's calculate the number of 4wheelers per 100 inhabitants.
for col in ['1990', '1991', '1992', '1993', '1994', '1995', '1996',
       '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005',
       '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']:
    df_4wheel[col + '_rel'] = (df_4wheel[col] / df_4wheel[col + '_pop'] * 100)
```

```python
# In which canton are the most 4-wheelers?
fig, ax = plt.subplots(figsize=(15,10))
df_4wheel[['2015_rel']].sort_values(by='2015_rel', ascending=True).plot(kind='barh', color='darkgreen', legend=None, ax=ax)
ax.set_title('Four Wheelers for the Folks in the Mountains')
ax.set_xlabel('Four Wheeler per 100 people')
ax.set_ylabel('')
plt.savefig('output/4wheelers.pdf')
```
