```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
gdpgrowth = pd.read_csv("GDP-growth.csv")
```

```python
gdpgrowth
```

```python
inflation = pd.read_csv("Inflation-rate.csv")
```

```python
inflation.drop(inflation.index[[6]])
```

```python
plt.style.use("fivethirtyeight")
```

```python
ax = gdpgrowth.plot(kind='line', x='gdp growth', y='year', label="gdp growth", linewidth=2)
inflation.plot(kind='line', x='year', y='Inflation rate',ax=ax, label='inflation rate', color='red',linewidth=2)

ax.set_title("Brazil's Recession")

ax.set_xlim((2010,2015))
ax.set_ylim((-6,12))

#plt.savefig("students ration.pdf", transparent = True)
```

```python
deficit = pd.read_csv("Budget-balance.csv")
new_deficit = deficit.drop(inflation.index[[6]])
```

```python
new_deficit
```

```python
ax = gdpgrowth.plot(kind='line', x='gdp growth', y='year', label="gdp growth(%)", color='#FF3C38', linewidth=2,figsize=(10,7))
inflation.plot(kind='line', x='year', y='Inflation rate',ax=ax, label='inflation rate(%)', color='#6699CC',linewidth=2)
#new_deficit.plot(kind='bar',x='Year', y='Budget balance', ax=ax, label='buget balance', color='orange')

ax.set_title("Brazil's Recession")

ax.set_xlim((2010,2015))
ax.set_ylim((-4,10))

#plt.savefig("Brazil's Recession.pdf")
```

```python
ax1 = new_deficit.plot(kind='bar',x='Year', y='Budget balance', label='buget balance(%GDP)', color='#FF8C42', figsize=(10,7))
ax1.set_ylim((-14,0))
plt.savefig("deficit.pdf")
```
