```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

%matplotlib inline
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
df = pd.read_excel("box_office_numbers.xlsx")
df
```

```python
#Certified Fresh

df[df['RottenTomatoes'] >= 75].count()
```

```python
fig, ax = plt.subplots(figsize=(15,10))
fig.set_facecolor('#B8D8BA')
ax.set_axis_bgcolor('#B8D8BA')

for category, group in df.groupby('Franchise'):
    ax.plot(group['Worldwide'], group['RottenTomatoes'], marker='o', linestyle='', label=category, 
            markeredgewidth=0, ms=10
        )

matplotlib.rcParams.update({'font.size': 18})
    
#ax.set_prop_cycle('color', ['#41337A', '#331E36'])    
#This doesn't work 

ax.set_xlim([0, 3000])
ax.set_ylim([0, 105])
ax.set_xlabel('Worldwide Box Office in Millions of Dollars')
ax.set_ylabel('Rating on Rotten Tomatoes')

ax.xaxis.grid(which='Major', color='grey', linestyle='-', linewidth=0.5) 
ax.yaxis.grid(which='Major', color='grey', linestyle='-', linewidth=0.5) 
ax.set_axisbelow(True)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#ax.spines['bottom'].set_visible(False)
#ax.spines['left'].set_visible(False)

plt.tick_params(
    top='off', #ticks along the top edge are off
    left='off', #ticks along the left edge are off
    right='off', #ticks along the right edge are off
    bottom='off', #ticks along the bottom edge are on
    labelright='off',
    labelleft='on',
    labeltop='on', #top label is on
    labelbottom='on' ) #bottom label is on

plt.savefig("movies.pdf", transparent=True)
```
