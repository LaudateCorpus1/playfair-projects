```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import dateutil.parser

%matplotlib inline
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
df = pd.read_excel("top_earning_actors_both.xlsx")
df
```

```python
fig, ax = plt.subplots(figsize=(15,10))
fig.set_facecolor('#e5f5f9')
ax.set_axis_bgcolor('#e5f5f9')

for category, group in df.groupby('Gender'):
    ax.plot(group['Earnings (Millions)'], group['Age (2015)'], marker='o', linestyle='', label=category, 
            markeredgewidth=0, ms=15
        )

matplotlib.rcParams.update({'font.size': 18})

ax.set_xlim([0, 85])
ax.set_ylim([0, 85])
ax.set_xlabel('Earnings (Millions)')
ax.set_ylabel('Age')

ax.xaxis.grid(which='Major', color='grey', linestyle='-', linewidth=0.5) 
ax.yaxis.grid(which='Major', color='grey', linestyle='-', linewidth=0.5) 
ax.set_axisbelow(True)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#ax.spines['bottom'].set_visible(False)
#ax.spines['left'].set_visible(False)

plt.tick_params(
    top='off', #ticks along the top edge are off
    left='off', #ticks along the left edge are off
    right='off', #ticks along the right edge are off
    bottom='off', #ticks along the bottom edge are on
    labelright='off',
    labelleft='on',
    labeltop='off', #top label is on
    labelbottom='on' ) #bottom label is on

plt.savefig("movie_stars_distribution.pdf", transparent=True)
```
