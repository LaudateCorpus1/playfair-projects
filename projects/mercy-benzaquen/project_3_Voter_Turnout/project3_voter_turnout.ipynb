```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
df= pd.read_excel("Data_voter_turnout.xlsx")
```

```python
df.tail()
```

```python
df.columns
```

```python
df.head()
```

```python
df.transpose
df
```

```python
df['new_col'] = df.index
```

```python
fig, ax = plt.subplots()

df.plot(kind='scatter',x='% of voting age population', y='new_col', ax=ax)
df.plot(kind='scatter',x='% of registered voters', y='new_col',color='green',ax=ax)
ax.xaxis.grid(which= 'major', color='grey', linestyle='-', linewidth=1)
ax.xaxis.grid(which= 'minor', color='grey', linestyle='-', linewidth=1)
ax.yaxis.grid(which= 'major', color='grey', linestyle='-', linewidth=1)
ax.yaxis.grid(which= 'minor', color='grey', linestyle='-', linewidth=1)
ax.set_ylim([40,-2])
ax.set_xlim([10,100])
ax.grid(linestyle='-')
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.tick_params(which = 'major', top= 'on',right='on', bottom= 'on', labeltop= 'on', labelbottom= 'on', labelleft= 'on')
plt.tick_params(which = 'minor', top= 'on',right='on', bottom= 'on', labeltop= 'on', labelbottom= 'on', labelleft= 'on')
ax.minorticks_on()

plt.savefig("turnout_chart_1.pdf",transparent=True)
```
