```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
df = pd.read_excel("horas-extra-no-pagadas-por-sexo.xls")
```

```python
df
```

```python
df['Total horas extra (000)'] = df['Horas hombres (000)'] + df['Horas mujeres (000)']
```

```python
df['Media Horas extra hombres'] = (df['Horas hombres (000)']/df['Ocupados hombres (000)'])*1000
```

```python
df['Media Horas extra mujeres'] = (df['Horas mujeres (000)']/df['Ocupados mujeres (000)'])*1000
```

```python
df
```

```python
media_horas_total = (df['Total horas extra (000)'].sum()/(df['Ocupados hombres (000)'].sum() + df['Ocupados mujeres (000)'].sum()))*1000
```

```python
media_horas_total
```

```python
media_horas_mujeres = (df['Horas mujeres (000)'].sum()/df['Ocupados mujeres (000)'].sum()) * 1000
```

```python
media_horas_mujeres
```

```python
media_horas_hombres = (df['Horas hombres (000)'].sum()/df['Ocupados hombres (000)'].sum()) * 1000
```

```python
media_horas_hombres
```

```python
(media_horas_mujeres/media_horas_hombres)*100
```

```python
df.sort_values(by='Ocupados mujeres (000)', ascending=False)
```

```python
df.index = df['Periodo']
```

```python
df.head()
```

```python
fig, ax = plt.subplots()

df.plot(y=['Media Horas extra hombres', 'Media Horas extra mujeres'], x='Periodo', ax=ax, legend=True, figsize=(15,10))

fig.set_facecolor('black')
ax.set_axis_bgcolor('black')

ax.tick_params(axis='x', colors='grey')
ax.tick_params(axis='y', colors='grey')

plt.xticks(x='Periodo', rotation='horizontal')
plt.margins(0.05)
plt.subplots_adjust(bottom=0.05)

plt.tick_params(
    which='both',
    color='white',
    top='off', 
    left='off',
    right='off',
    bottom='on',
    labelright='off', 
    labeltop='off',
    labelbottom='on')

ax.set_title("Spanish workers do 3 hours of unpaid overtime a day")
ax.set_xlabel("Period")
ax.set_ylabel("Hours per quarter")
ax.xaxis.label.set_color('gray')
ax.yaxis.label.set_color('gray')
ax.title.set_color('gray')

plt.savefig("median-overtime-gender.pdf")
```
