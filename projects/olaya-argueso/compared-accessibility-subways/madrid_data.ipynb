## Accessibility of Madrid's subway stations

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
madrid_df = pd.read_excel("estaciones-metro-madrid.xlsx")
madrid_df.head()
```

**1) How many stations are completely inaccessible, not having either elevator or escalators?**

```python
estaciones_sin_ascensor_ni_escaleras = madrid_df[(madrid_df['Escaleras mecánicas'] == 0) & (madrid_df['Ascensor'] == 0) & (madrid_df['Accesible a personas con discapacidad'] == 0)]
```

## Visualisation

```python
viz_df = pd.pivot_table(madrid_df, values=['Accesible a personas con discapacidad', 'Escaleras mecánicas', 'Ascensor'], index=['Línea 1']) 
```

```python
viz_df
```

```python
viz_df['Line name'] = viz_df.index
```

```python
viz_df.head()
```

```python
new_viz_df = viz_df.reset_index()
```

```python
new_viz_df
```

```python
new_viz_df['% of accessible stations per line'] = round(new_viz_df['Accesible a personas con discapacidad'] * 100)
```

```python
new_viz_df['% of not accessible stations per line'] = 100 - new_viz_df['% of accessible stations per line']
```

```python
new_viz_df
```

```python
fig, ax = plt.subplots()

ax = new_viz_df.plot(kind='bar', x='Line name', y=['% of not accessible stations per line', '% of accessible stations per line'], stacked=True, rot=1, ax=ax, legend=True, width=0.80, figsize=(10,5))

ax.yaxis.grid(which='major', color='grey', linestyle='-', linewidth=0.5)

fig.set_facecolor('gainsboro')
ax.patch.set_alpha(0.15)
ax.set_axis_bgcolor('gainsboro')

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.set_title('How accessible are Madrid\'s subway stations?')
ax.set_ylabel('% stations, according to their accessibility')

plt.savefig("accessiblity-per-line.pdf")
```
