## Accessibility of Madrid's subway stations

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
madrid_df = pd.read_excel("estaciones-metro-madrid.xlsx")
madrid_df.head()
```

**1) How many stations are completely inaccessible, not having either elevator or escalators?**

```python
estaciones_sin_ascensor_ni_escaleras = madrid_df[(madrid_df['Escaleras mecánicas'] == 0) & (madrid_df['Ascensor'] == 0) & (madrid_df['Accesible a personas con discapacidad'] == 0)]
```

```python
len(estaciones_sin_ascensor_ni_escaleras)
```

**Answer**: there are 34.

```python
estaciones_sin_ascensor_ni_escaleras
```

**2) To which lines do these stations belong?**

```python
estaciones_sin_ascensor_ni_escaleras.sort_values(by='Línea 1')
```

**Answer**: Mainly to lines 1, 2, and 5.

```python
(estaciones_sin_ascensor_ni_escaleras['Línea 1'] == 1).value_counts()
```

**A third (11) belong to Line 1**, which is the oldest (1919), and is currently (summer 2016) undergoing major construction works, which affect all these 11 non-accessible stations. The works do not include making them accessible, though.

```python
(estaciones_sin_ascensor_ni_escaleras['Línea 1'] == 2).value_counts()
```

Seven of them, **(20%) belong to Line 2**, the second oldest.

```python
(estaciones_sin_ascensor_ni_escaleras['Línea 1'] == 5).value_counts()
```

**Another third (11) belong to Line 5**.

## Visualisation

```python
viz_df = pd.pivot_table(madrid_df, values=['Accesible a personas con discapacidad', 'Escaleras mecánicas', 'Ascensor'], index=['Línea 1']) 
```

```python
viz_df
```

```python
fig, ax = plt.subplots()

ax = viz_df.plot(kind='bar', width=0.80, figsize=(15,10), ax=ax)

ax.yaxis.grid(which='major', color='grey', linestyle='-', linewidth=0.5)

fig.set_facecolor('gainsboro')
ax.patch.set_alpha(0.15)
ax.set_axis_bgcolor('gainsboro')

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.set_title('How accessible are Madrid\'s subway stations?')
ax.set_ylabel('% stations, according to their accessibility')

plt.savefig("accessiblity-per-line.pdf")
```
